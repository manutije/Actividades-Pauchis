```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f(x) = -2x^2 +4x + 5 {h:1, k:7}
# x1 (-0.87) 
# x2 ( 2.87)
x = np.linspace(-6, 8, 500)
y = (-2*(x*x) +(4*x) + 5)
```

```python
fig,ax = plt.subplots()
ax.set_title('f(x) = -2x^2 +4x + 5 Rivera Arroyo Paula')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.plot(x,y,'-')
ax.plot(-0.87,0,'*')
ax.plot(2.87,0,'*')
ax.plot(1,7,'o')
ax.grid()
```

```python
# f(x) = 3x^2 -5x + 8 {h: 0.83, k: 5.92}
# x1 (inf) 
# x2 ( inf)
x = np.linspace(-6, 8, 500)
y = (3*(x*x) -(5*x) + 8)
```

```python
fig,ax = plt.subplots()
ax.set_title('f(x) = 3x^2 -5x + 8 Rivera Arroyo Paula')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.plot(x,y,'-')
ax.plot(0.83,5.92,'o')
ax.grid()
```

```python
# f(x) = -2.5x^2 -2.6x - 3.4 {h: -0.52, k: -2.73}
# x1 (inf) 
# x2 (inf)
x = np.linspace(-7, 6, 500)
y = (-2.5*(x*x) -(2.6*x) -3.4)
```
