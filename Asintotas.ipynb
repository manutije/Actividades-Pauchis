```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-50,50)
```

```python
f1 = 4/(5*x)
f2 = -6/((4*x)-1)
f3 = ((4*x)-2)/(3*x)
f4 = ((-7*x)+1)/((8*x)-3)
```

```python
fig,ax = plt.subplots()
ax.set_title('f(x) = 4/5x \n Rivera Arroyo Paula')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.plot(x,f1,'-')
ax.grid()
```

```python
fig,ax = plt.subplots()
ax.set_title('f(x) = -6/(4x - 1) \n Rivera Arroyo Paula')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.plot(x,f2,'-')
ax.grid()
```

```python
fig,ax = plt.subplots()
ax.set_title('f(x) = (4x - 2)/3x \n Rivera Arroyo Paula')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.plot(x,f3,'-')
ax.grid()
```

```python
fig,ax = plt.subplots()
ax.set_title('f(x) = (-7x + 1)/(8x - 3) \n Rivera Arroyo Paula')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.plot(x,f4,'-')
ax.grid()
```
