```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-50,50,500)
y = (7*(x*x)) - (5*x) + 4
```

```python
fig,ax = plt.subplots()
ax.set_title('A) f(x) = 7x^2 -5x + 4')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.plot(x,y,'-')
ax.grid()
```

```python
x = np.linspace(-50,50,500)
y = (-4*(x*x)) + (6*x) - 8
```

```python
fig,ax = plt.subplots()
ax.set_title('B) f(x) = -4x^2 +6x - 8')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.plot(x,y,'-')
ax.grid()
```
